```python
# Assuming your data arrays are pre-loaded and NaN values are filtered out as described
control = ph_his_np.flatten()[~np.isnan(ph_his_np.flatten())]
test1 = ph_ssp119_np.flatten()[~np.isnan(ph_ssp119_np.flatten())]
test2 = ph_ssp126_np.flatten()[~np.isnan(ph_ssp126_np.flatten())]
test3 = ph_ssp245_np.flatten()[~np.isnan(ph_ssp245_np.flatten())]
test4 = ph_ssp370_np.flatten()[~np.isnan(ph_ssp370_np.flatten())]
test5 = ph_ssp585_np.flatten()[~np.isnan(ph_ssp585_np.flatten())]

# Concatenate all valid data points
data = np.concatenate([control, test1, test2, test3, test4, test5])
groups = np.concatenate([
    np.zeros(len(control)),
    np.ones(len(test1)),
    2*np.ones(len(test2)),
    3*np.ones(len(test3)),
    4*np.ones(len(test4)),
    5*np.ones(len(test5))
])

# Perform Kruskal-Wallis test
stat, p_value = stats.kruskal(control, test1, test2, test3, test4, test5)
print("Kruskal-Wallis Test Statistic:", stat)
print("P-value:", p_value)

# If significant, perform Dunn's test
if p_value < 0.05:
    print("\nProceeding with Dunn's post-hoc test due to significant Kruskal-Wallis result.\n")
    p_values_matrix = sp.posthoc_dunn([control, test1, test2, test3, test4, test5], p_adjust='bonferroni')
    print("Dunn's Test pairwise p-values with Bonferroni correction:")
    print(p_values_matrix)
else:
    print("No significant differences detected; no further testing needed.")
```
